# Titanic Survival Analysis
### Python Data Cleaning + Exploratory Data Analysis

Objective:
Analyze Titanic passenger data to identify major factors affecting survival using data cleaning, feature engineering, and exploratory data analysis.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"Titanic.csv")
df
```

```python
df.isna().sum()
```

```python
df.drop(columns="Cabin",inplace=True)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df["Embarked"] = df["Embarked"].map({
    "C": "Cherbourg",
    "Q": "Queenstown",
    "S": "Southampton"
})
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df["Title"] = df["Name"].str.extract(" ([A-Za-z]+)\.",expand=False)
```

```python
df.drop(columns="Name",inplace=True)
```

```python
df["TicketPrefix"] = df["Ticket"].str.extract(r'([A-Za-z./ ]+)', expand=False)
```

```python
df["TicketPrefix"] = df["TicketPrefix"].fillna("Numeric")
```

```python
df.groupby("Sex")["Survived"].mean()*100
```

```python
df.groupby("Pclass")["Survived"].mean()*100
```

```python
df["Ticket_Class"] = np.where(df["Fare"] > 100 ,"Luxury",np.where(df["Fare"] > 50,"Premium","General"))
```

```python
df.groupby("Ticket_Class")["Survived"].mean()*100
```

```python
df.drop(columns="Ticket_Class")
```

```python
df.groupby(["Pclass","Sex"])["Survived"].mean().unstack(fill_value=0)*100
```

```python
df.groupby("Embarked")["Survived"].mean()*100
```

```python
df.to_excel("Titanic_cleaned.xlsx", index=False)
```

## Key Insights
- Female passengers had higher survival rates
- First-class passengers had better survival chances
- First-class females had highest survival probability
- Third-class males had lowest survival probability
- Higher fare passengers survived more frequently
- Small family groups showed better survival outcomes
